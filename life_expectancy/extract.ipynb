```python
from bs4 import BeautifulSoup
import requests
import pandas as pd


html_text=requests.get('https://en.wikipedia.org/wiki/List_of_Indian_states_by_life_expectancy_at_birth')
soup=BeautifulSoup(html_text.content,'html.parser')
data_table=soup.find('table',{'class':'wikitable sortable'})

print(data_table)
```

```python
df=pd.read_html(str(data_table))

# convert list to dataframe
df=pd.DataFrame(df[0])
print(df.head(2))
df.to_csv('life.csv',header=None,index=False)

```

```python
from fuzzywuzzy import fuzz
list1=['Kerala', 'Jammu and Kashmir', 'Himachal Pradesh', 'Punjab', 'Maharashtra', 'Uttarakhand', 'Tamil Nadu', 'West Bengal', 'Gujarat', 'Haryana', 'Karnataka', 'Andhra Pradesh (includes Telangana)', 'Delhi', 'Bihar', 'Jharkhand', 'Rajasthan', 'Odisha', 'Assam', 'Madhya Pradesh', 'Chhattisgarh', 'Uttar Pradesh', 'Goa', 'Sikkim', 'Manipur', 'Tripura', 'Arunachal Pradesh', 'Nagaland', 'Meghalaya', 'Telangana', 'Mizoram', 'Andaman and Nicobar Islands', 'Puducherry', 'Lakshadweep', 'Chandigarh', 'Daman and Diu and Dadra nagar haveli']

list2=['Andaman & Nicobar Island', 'Andhra', 'Arunanchal', 'Assam', 'Bengal', 'Bihar', 'Chandigarh', 'Chhattisgarh', 'Dadra & Nagar Haveli and Daman & Diu', 'Delhi', 'Goa', 'Gujarat', 'Haryana', 'Himachal', 'Jammu & Kashmir', 'Jharkhand', 'Karnataka', 'Kerala', 'Lakshadweep', 'Madhya Pradesh', 'Maharashtra', 'Manipur', 'Meghalaya', 'Mizoram', 'Nagaland', 'Odisha', 'Puducherry', 'Punjab', 'Rajasthan', 'Sikkim', 'Tamil Nadu', 'Telangana', 'Tripura', 'Uttar Pradesh', 'Uttarakhand','Ladakh']


list3=[]
for name1 in list1:
    match_found = False
    for name2 in list2:
        similarity_ratio = fuzz.ratio(name1, name2)
        partial_ratio = fuzz.partial_ratio(name1, name2)
        
        if 'diu' in name1.lower() and 'diu' in name2.lower():
            similarity_ratio=85
        
        if 'island' in name1.lower() and 'island' in name2.lower() :
            similarity_ratio=85
            
        if 'orissa' in name1.lower() and 'odisha' in name2.lower():
            similarity_ratio=85
            
       

        # Check for matches based on multiple criteria and thresholds:
        if similarity_ratio >= 85 or partial_ratio >= 85:  # Prioritize high similarity
            print(f"{name1} from list1  matches {name2} from list2 with:{similarity_ratio}")
            name1=name2
            list3.append(name2)
            match_found = True
            
        elif similarity_ratio >= 75:  # Consider slightly lower matches
            print(f"{name1} from list1 matches <b>{name2}</b> from list2 with: {similarity_ratio} but potentially")
            list3.append(name2)
            match_found = True
            
    if not match_found:
        print(f"No close match found for {name1} "*2)

            
print(len(list1))
print(len(list2))
print(len(list3))
print(list3)
```

```python
items_not_in_list3=[]
if len(list1)==len(list3) and len(list1) < len(list2):
    items_not_in_list3 = [item for item in list2 if item not in list3]

print("Items in list2 but not in list3:")
print(items_not_in_list3)

for i, name in enumerate(list3):
    if name == "Jammu & Kashmir":
        list3[i] = 'J&K'
    elif name == 'Dadra & Nagar Haveli and Daman & Diu':
        list3[i]='DDNH'

        
list3
```
